## Wordnet

```python
database= "wordnet"
```

```python
URI = "neo4j+s://demo.neo4jlabs.com"
AUTH = (database, database)
```

```python
query = '''MATCH (f:ontolex__Form)<-[:ontolex__canonicalForm]-(e:ontolex__LexicalEntry)-[:ontolex__sense]->(s:ontolex__LexicalSense)-[:ontolex__isLexicalizedSenseOf]->(c:ontolex__LexicalConcept) RETURN f,e,s,c'''
```

```python
def make_query_safe(query,n=100):
        str_n = str(n)
        parsed_query = query.split(' LIMIT ')
        if len(parsed_query) == 1:
            modified_query = ' LIMIT '.join([parsed_query[0],str_n ])
        else:
            modified_query = ' LIMIT '.join([parsed_query[:-1],str_n ])
        return modified_query           
```

```python
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()
    print("Connection established.")
    records, summary, keys  = driver.execute_query(make_query_safe(query),database=database)
```
